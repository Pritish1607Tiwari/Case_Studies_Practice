```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
customers=pd.read_csv("Customer.csv")
products=pd.read_csv("prod_cat_info.csv")
transactions=pd.read_csv("Transactions.csv")

print(customers)
print(products)
print(transactions)
```

# Q1
Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure to
keep all customers who have done transactions with us and select the join type accordingly.


```python
temp = pd.merge(left=transactions,right=customers,left_on="cust_id",right_on="customer_Id",how="inner")
```

```python
customer_final = pd.merge(left=temp,right=products,
         left_on=["prod_cat_code","prod_subcat_code"],
         right_on=["prod_cat_code","prod_sub_cat_code"],
         how="inner")
```

# Q6
 Which City code has the maximum customers and what was the percentage of customers from 
that city?

```python
customers
```

```python
cust_count=pd.pivot_table(data=customers,
               index="city_code",
               values="customer_Id",
               aggfunc="count"
               )
max_count_cust=cust_count.sort_values(by="customer_Id",ascending=False).iloc[0:1,:]


city_code_max_cust= max_count_cust.index[0]
total_cust=max_count_cust.iloc[0,0]

print(f"Max Customer are from city Code = {city_code_max_cust} , with total customer = {total_cust}")
percentage_cust= (total_cust/len(customers))*100
print(f"% of people in city code {city_code_max_cust} = {round(percentage_cust)}%")
```

 # Q11 | A
 For all customers aged between 25 - 35, find out:
 What was the total amount spent for “Electronics” and “Books” product categories?

```python
customer_final.DOB = pd.to_datetime(customer_final.DOB)
```

```python
customer_final["AGE"]=(dt.datetime.now()- customer_final.DOB).dt.days // 365.25
```

```python
df = customer_final[(customer_final.AGE.between(23,35))
                & customer_final.prod_cat.isin(["Electronics","Books"])]
sum = round(df["total_amt"].sum())
print(" total amount spent for “Electronics” and “Books” product categories b/w ages 25 and 35 = ",sum)
```

# B
What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?


```python
DF = customer_final[(customer_final.tran_date > "2014-01-01")
    & (customer_final.tran_date < "2014-03-01")]
sum = round(DF["total_amt"].sum())
print("total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014 = ", sum)
```
